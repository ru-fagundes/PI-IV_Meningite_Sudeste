## üìä Verificar Arquivos Gerados

```python
import os
import pandas as pd

print("="*80)
print("VERIFICA√á√ÉO DE ARQUIVOS GERADOS")
print("="*80)

# Verificar data/raw/
print("\nüìÅ data/raw/")
raw_files = os.listdir('data/raw')
for i, file in enumerate(sorted(raw_files), 1):
    size = os.path.getsize(f'data/raw/{file}') / 1024
    print(f"   {i}. {file:<35} ({size:.2f} KB)")

# Verificar data/processed/
print("\nüìÅ data/processed/")
processed_files = os.listdir('data/processed')
for i, file in enumerate(sorted(processed_files), 1):
    size = os.path.getsize(f'data/processed/{file}') / 1024
    print(f"   {i}. {file:<35} ({size:.2f} KB)")

# Verificar data/powerbi/
print("\nüìÅ data/powerbi/")
if os.path.exists('data/powerbi'):
    powerbi_files = os.listdir('data/powerbi')
    for i, file in enumerate(sorted(powerbi_files), 1):
        size = os.path.getsize(f'data/powerbi/{file}') / 1024
        print(f"   {i}. {file:<35} ({size:.2f} KB)")
else:
    print("   ‚ö†Ô∏è Pasta n√£o existe ainda (ser√° criada na an√°lise ML)")

print("\n" + "="*80)
print("‚úÖ VERIFICA√á√ÉO CONCLU√çDA")
print("="*80)
```
